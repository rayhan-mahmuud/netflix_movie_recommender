```python
movie = df[df.type=="Movie"] 
movie.head(5)
```

```python
movie.shape
```

```python
movie = movie.drop(columns = ["date_added", "release_year","duration","type","rating", "show_id"])
movie.head()
```

```python
movie.isnull().sum()
```

```python
movie.shape
```

```python
movie = movie.dropna()
movie.head()
```

```python
movie.shape
```

```python
movie["tags"] = movie["director"]+" "+movie["cast"]+" "+movie["country"]+" "+movie["listed_in"]+" "+movie["description"]
movie.head()
```

```python
movie = movie.reset_index()
movie = movie.rename(columns = {"index":"movie_id"})
movie.head()
```

```python
movies = movie.drop(columns = ["director","cast","country","listed_in","description"],axis = 1)
movies.to_csv("G:/Python/ML/Projects/Movie Recommender/data/movies.csv")
```

```python
movies.head()
```

```python
movies.tags[0]
```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
porter = PorterStemmer()
```

```python
def stemmer(tags):
    stemmed = []
    tokens = word_tokenize(tags)
    for token in tokens:
        stemmed.append(porter.stem(token))
    return " ".join(stemmed)    
    
```

```python
movies["tags"] = movies["tags"].apply(stemmer)
movies.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA 
```

```python
tfidf = TfidfVectorizer()
vectors = tfidf.fit_transform(movies["tags"])
vectors = vectors.toarray()
vectors
```

```python
vectors.shape
```

```python
pca = PCA(0.9)
vectors_pca = pca.fit_transform(vectors)
vectors_pca
```

```python
vectors_pca.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_of_movies = cosine_similarity(vectors_pca)
similarity_of_movies.shape
```

```python
similarity_of_movies[0]
```

```python
movies.title[0]
```

```python
test_sim = sorted(enumerate(similarity_of_movies[0]), reverse=True, key = lambda x: x[1])
test_sim
```

```python
def recommender(movie):
    result = []
    count = 0
    sim = similarity_of_movies[movies[movies["title"]==movie].index[0]]
    sorted_sim = sorted(enumerate(sim), reverse=True, key = lambda x: x[1])
    for index, val in sorted_sim:
        result.append(movies.title[index])
        count+=1
        if count>8:
            break
    return result[1:]
```

```python
recommender("Jeans")
```

```python
recommender("Grown Ups")
```
